# Studing and clearing dataset

First of all connect all libraries for work.

```python
#import libraries for work
import numpy as np
import pandas as pd
import sqlite3

#for better visibility
from IPython.display import display, HTML
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
#connect to the created Database
conn = sqlite3.connect("movies_database.sqlite")
cur = conn.cursor()
```

```python
#information about tables
    #select all tables from database
cur.execute("""select name from sqlite_master where type = 'table'""")
    
    # put all tables name into variable tables_name
tables_name = cur.fetchall()
    
    # for each table check information about columns
for table_name in tables_name:
    df = pd.read_sql(f"PRAGMA table_info("+(table_name[0])+")", conn)
    
    # count rows in each table
    row_count = cur.execute("select count(*) from "+(table_name[0])+"").fetchone()[0]
        
    # pretty printing
    print(color.BLUE+color.BOLD + table_name[0]+ color.END, "==", row_count, "rows")
    display(HTML(df.to_html()))
```

# How we can see on this results: tables from different sources have different key set to determine the uniqueness of a record

```python
# Quick look at the tables from my database
cur.execute("""select name from sqlite_master where type = 'table'""")
tables_name = cur.fetchall()

for table_name in tables_name:
    print(f"Looking at Table: ", color.BLUE+color.BOLD + table_name[0]+ color.END)
    df = pd.read_sql(f"select * from [{table_name[0]}]", conn)
    # check for dublicate rows
    x = df.duplicated().sum()
    #df.isna().sum() - check for missing values missing values.
    print(f"Looking at Table for duplicate rows: ", x)
    print(f"Checking for missing values: ", "\n", df.isna().sum())
    print(f"Total rows: ",len(df.index))
    display(df.head())
    
    print("\n")
```

## Check and convert all money columns into integers

We have a few tables with budgeted value, but we cannot work with them faster, because all values has differents type.

```python
cur.execute("SELECT * FROM tn_movie_budgets")
movie_budgets = pd.DataFrame(cur.fetchall())
movie_budgets.columns = [x[0] for x in cur.description]
movie_budgets.head()
```

```python
# check missing values
print(movie_budgets.isna().sum())
print(movie_budgets.shape) 
```

```python
movie_budgets['production_budget'] = movie_budgets['production_budget'].str.replace("$","").str.replace(",","").astype('int')
```

```python
#SUBSTRING ( expression ,start , length )  
cur.execute('''SELECT movie, 
                CAST(REPLACE(SUBSTR(production_budget,2),',','')as INTEGER) as production_budget, 
                CAST(REPLACE(SUBSTR(domestic_gross,2),',','')as INTEGER) as domestic_gross, 
                CAST(REPLACE(SUBSTR(worldwide_gross,2),',','')as INTEGER) as worldwide_gross 
                FROM tn_movie_budgets''')
df2 = pd.DataFrame(cur.fetchall())
df2.columns = [x[0] for x in cur.description]
df2.head()
```

```python
movie_budgets['domestic_gross'] = movie_budgets['domestic_gross'].str.replace("$","").str.replace(",","").astype('int')
```

```python
movie_budgets['worldwide_gross'] = movie_budgets['worldwide_gross'].str.replace("$","").str.replace(",","").astype('int')
movie_budgets.info()
```
